```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('po1_data.txt')
```

```python
df.columns = ['Subject identifier', 'Jitter %', 'Jitter ms', 'Jitter r.a.p', 'Jitter p.p.q.5', 'Jitter d.d.p',
             'Shimmer %', 'Shimmer dB', 'Shimmer a.p.q.3', 'Shimmer a.p.q.5', 'Shimmer a.p.q.11', 'Shimmer d.d.a',
             'Harmonicity (autocorrelation)', 'Harmonicity NHR', 'Harmonicity HNR', 'Median Pitch', 'Mean Pitch', 
              'SD Pitch', 'Minimum Pitch', 'Maximum Pitch', 'Pulse (no)', 'Pulse (No. of periods)', 
              'Pulse (mean period)', 'Pulse (SD period)', 'Voice (unframed voice)', 'Voice (no. of breaks)', 
             'Voice (degree of breaks)', 'UPDRS', 'PD indicator']
df.head(30)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
data = df

# Subset the dataset into PD and healthy groups
pd_group = data[data['PD indicator'] == 1]
healthy_group = data[data['PD indicator'] == 0]

# Define the columns related to jitter and shimmer measures
jitter_columns = ['Jitter %', 'Jitter ms', 'Jitter r.a.p', 'Jitter p.p.q.5', 'Jitter d.d.p']
shimmer_columns = ['Shimmer %', 'Shimmer dB', 'Shimmer a.p.q.3', 'Shimmer a.p.q.5', 'Shimmer a.p.q.11', 'Shimmer d.d.a']

# Plot histograms for jitter measures
plt.figure(figsize=(12, 6))
for column in jitter_columns:
    plt.hist(pd_group[column], alpha=0.5, label='PD')
    plt.hist(healthy_group[column], alpha=0.5, label='Healthy')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.legend()
    plt.title('Histograms of Jitter Measures')
    plt.show()

# Create box plots for shimmer measures
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='PD indicator', y=shimmer_columns[0])
plt.ylabel('Shimmer %')
plt.title('Box Plot of Shimmer %')
plt.show()

# Create density plots for shimmer measures
plt.figure(figsize=(12, 6))
for column in shimmer_columns:
    sns.kdeplot(pd_group[column], label='PD')
    sns.kdeplot(healthy_group[column], label='Healthy')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.legend()
    plt.title('Density Plots of Shimmer Measures')
    plt.show()
```

```python
# Load your dataset
data = df
print('This is the Dataframe:')
print(data)

data['PD indicator'] = data['PD indicator'].replace({0: 'Healthy', 1: 'PD'})

# Define the columns related to jitter, shimmer, pitch, and pulse measures
jitter_columns = ['Jitter %', 'Jitter ms', 'Jitter r.a.p', 'Jitter p.p.q.5', 'Jitter d.d.p']
shimmer_columns = ['Shimmer %', 'Shimmer dB', 'Shimmer a.p.q.3', 'Shimmer a.p.q.5', 'Shimmer a.p.q.11', 'Shimmer d.d.a']
pitch_columns = ['Median Pitch', 'Mean Pitch', 'SD Pitch', 'Minimum Pitch', 'Maximum Pitch']
pulse_columns = ['Pulse (no)', 'Pulse (No. of periods)', 'Pulse (mean period)', 'Pulse (SD period)']
voice_columns = ['Voice (unframed voice)', 'Voice (no. of breaks)', 'Voice (degree of breaks)']


# Calculate statistics for each measure based on PD indicator
grouped_data = data.groupby('PD indicator')
jitter_stats = grouped_data[jitter_columns].describe()
shimmer_stats = grouped_data[shimmer_columns].describe()
pitch_stats = grouped_data[pitch_columns].describe()
pulse_stats = grouped_data[pulse_columns].describe()
voice_stats = grouped_data[voice_columns].describe()

# Print the statistics for each measure based on PD indicator
print("Jitter Statistics:")
print(jitter_stats)

print("\nShimmer Statistics:")
print(shimmer_stats)

print("\nPitch Statistics:")
print(pitch_stats)

print("\nPulse Statistics:")
print(pulse_stats)

print("\nVocie Statistics:")
print(voice_stats)

# print('This is jitter DF')
jitterDf = pd.DataFrame(jitter_stats)
sorted_jitterDf = jitterDf.sort_values(by="PD indicator")
sorted_jitterDf.to_csv('/Users/beck/CDU/DataScience/Project 1/jitter.csv', index=True, header=True)

# print('This is Shimmer DF')
shimmerDf = pd.DataFrame(shimmer_stats)
sorted_shimmerDf = shimmerDf.sort_values(by="PD indicator")
sorted_shimmerDf.to_csv('/Users/beck/CDU/DataScience/Project 1/shimmer.csv', index=True, header=True)

# print('This is Pitch DF')
pitchDf = pd.DataFrame(pitch_stats)
sorted_pitchDf = pitchDf.sort_values(by="PD indicator")
sorted_pitchDf.to_csv('/Users/beck/CDU/DataScience/Project 1/pitch.csv', index=True, header=True)

# print('This is Pulse DF')
pulseDf = pd.DataFrame(pulse_stats)
sorted_pulseDf = pulseDf.sort_values(by="PD indicator")
sorted_pulseDf.to_csv('/Users/beck/CDU/DataScience/Project 1/pulse.csv', index=True, header=True)

# print('This is Voice DF')
voiceDf = pd.DataFrame(voice_stats)
sorted_voiceDf = voiceDf.sort_values(by="PD indicator")
sorted_voiceDf.to_csv('/Users/beck/CDU/DataScience/Project 1/voice.csv', index=True, header=True)


# show the coffiecient correlation
# chose the highest relevant variables
```
